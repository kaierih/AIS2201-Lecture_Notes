# Digitale filtre til nå:

* **FIR** - **F**inite **I**mpulse **R**esponse
* Fellesbetegnelse for filter som kalkulerer utgangsveriden $y[n]$ *kun* basert på en lineær kombinasjon av de $M+1$ nyeste verdiene til inngangssignalet $x[n]$ ved forskjellige sampletidspunkt, der $M$ er filterets filterorden.
$$y[n] = \sum_{k=0}^{M} b_k \cdot x[n-k]$$
* Impulsresponsen $h[n]$ til et FIR filter gjengir filterkoeffisientene $b_k$
$$h[n] = \sum_{k=0}^{M} b_k \cdot \delta[n-k] = 
\begin{cases}
b_n, \ \ 0\leq n \leq M\\
0, \ \ \ \ \text{ellers}
\end{cases}$$
* Filtrereing av et signal $x[n]$ tilsvarer å utføre konvolusjon mellom signalet og impulsresponsen $h[n]$.
$$y[n] = x[n]*h[n]$$

1. Hva er filterets impulsrespons $h[n]$?
2. Hva er filterets stegrespons $s[n]$?

1. impulsrespons $h[n]$:
    * Vi finner impulsresponsen ved å sette inngangssignalet $x[n] = \delta[n]$, og regner ut ulike verdier for $y[n]$:

| $n$ | $x[n] = \delta[n]$ | $y[n] = 0.1\cdot x[n] + y[n-1]$ |
|---|---|---|
|$0$|$1$|$0.1 \cdot 1 + 0.9 \cdot 0 = 0.1$|
|$1$|$0$|$0.1 \cdot 0 + 0.9 \cdot 0.1 = 0.09$|
|$2$|$0$|$0.1 \cdot 0 + 0.9 \cdot 0.09 = 0.081$|
|$3$|$0$|$0.1 \cdot 0 + 0.9 \cdot 0.081 = 0.0729$|
|$\vdots$|$\vdots$|$\vdots$|
|$n$|$\delta[n]$|$0.1 \cdot 0.9 ^ n \cdot u[n]$|

2. stegrespons $s[n]$:
\begin{align}
s[n] &= h[n]*u[n]\\
&= \sum_{k=0}^{\infty} 0.1\cdot 0.9^{n} \cdot u[n-k] \\
&= 0.1 \cdot \sum_{k=0}^{n} 0.9^n \\
&= 0.1 \cdot \frac{1-0.9^{n+1}}{1-0.9} \cdot u[n]\\ 
&= \frac{0.1}{0.1} \left( 1 - 0.9^{n+1} \right) \cdot u[n]\\
&= \left( 1 - 0.9^{n+1} \right) \cdot u[n]\\
\end{align}
* Viktig detalj: 
    * Dersom denne stegresponsen forsinkes med 1 sample, får vi et uttrykk som korresponderer med en samplet stegrespons fra et førsteordens *kontinuerlig* LTI-system.

## Mer om stegrespons *(sprangrespons)*

* Synliggjør filterets "responstid". 
* IIR-filtre vil som regel konvergere mot en stasjonær verdi.

## Kommentarer til filteret $y[n] = 0.1\cdot x[n] + 0.9\cdot y[n-1]$:

1. Et slikt filter kalles ofte et "eksponentielt glattefilter" da det er et lavpassfilter og impulsresponsen er et eksponentialuttrykk. Et slikt filter vil ofte oppgis med følgende generelle formel:
$$y[n] = \alpha\cdot x[n] + (1-\alpha) \cdot y[n-1], \ \ \ 0 <\alpha < 1$$ 


2. Stegresponsen til filteret er nesten en diskrét utgave av stegresponsen til et førsteordens analogt lavpassfilter. 

# IIR filtre generell form

* **IIR** - **I**nfinite **I**mpulse **R**esponse
* Fellesbetegnelse for filter som kalkulerer utgangsveriden $y[n]$ basert på en lineær kombinasjon av verdiene til inngangssignalet ved foregående sampletidspunkt $x[n-k], \ \ k\in \{0, 1, 2, \ldots\}$ ***og*** tidligere kalkulerte verdier til utgangssignalet $y[n-k], \ \ k \in \{1, 2, \ldots\}$.
    * Har en tydelig parallell til analoge filtre.
    * Har to sett med koeffisienter: $a_k$ og $b_k$.


* **Differanseligning:**
$$\sum_{k=0}^{N} a_k \cdot y[n-k] = \sum_{k=0}^{M} b_k \cdot x[n-k]$$
* **Filteralgoritme:**
$$y[n] = \frac{1}{a_0}\cdot \left(\sum_{k=0}^{M} b_k \cdot x[n-k] - \sum_{k=1}^{N}a_k\cdot  y[n-k] \right)$$

## Mellomregning differanseligning $\rightarrow$ filteralgoritme
\begin{align}
\sum_{k=0}^{N} a_k \cdot y[n-k] &= \sum_{k=0}^{M} b_k \cdot x[n-k]\\
a_0\cdot y[n]+\sum_{k=1}^{N} a_k \cdot y[n-k] &= \sum_{k=0}^{M} b_k \cdot x[n-k]\\
a_0\cdot y[n] &= \sum_{k=0}^{M} b_k \cdot x[n-k]-\sum_{k=1}^{N} a_k \cdot y[n-k]\\
y[n] &=\frac{1}{a_0}\cdot \left(\sum_{k=0}^{M} b_k \cdot x[n-k] - \sum_{k=1}^{N}a_k\cdot  y[n-k] \right)
\end{align}

* Ettersom impulsresponsen har uendelig varighet, er det ikke hensiktsmessig utføre filtrering ved å regne ut konvolusjon numerisk. Da bruker man heller filteralgoritmen, og finner utgangssamplene i kronologisk rekkefølge.

## Regneeksempel 2:

* Finn *filterkoeffisientene* $a_k$ og $b_k$ for filteret $y[n] = 0.1\cdot x[n] + 0.9\cdot y[n-1]$.

###  Løsning:

1. Balanser uttrykket til standardform for differanseligning:
$$1 \cdot y[n] - 0.9\cdot y[n-1] = 0.1\cdot x[n] $$
2. Koeffisientene $a_k$ vil være konstanten som skalerer $y[n-k]$, og koeffisisentene $ b_k$ vil være konstanten som skalerer $x[n-k]$:
$$\begin{align}
b_0 &= 0.1 \\
a_0 &= 1 \\
a_1 &= -0.9
\end{align}$$
    * Alle andre koeffisienter vil ha verdien 0.

## Kodeeksempel 1:

* Lag en funksjon i `Python` for å utføre filtrering av et signal gitt parametrene `b`, `a` og `x`.
* Filteralgoritme:
$$y[n] =\frac{1}{a_0}\cdot \left(\sum_{k=0}^{M} b_k \cdot x[n-k] - \sum_{k=1}^{N}a_k\cdot  y[n-k] \right)$$

```python
def IIRfilt(b, a, x):
    M = len(b)-1 # Filterorden FIR "del"
    N = len(a)-1 # Filterorden IIR "del"
    L = len(x)   # Lengde på inngangssignal
    
    y = np.zeros(L) # Iniatilisering av utgangssignal
    
    for n in range(L):
        # For-løkke med første summasjonstegn
        for k in range(min(M+1, n+1)):
            y[n] += b[k]*x[n-k]
            
        # For-løkke med andre summasjonstegn
        for k in range(1, min(N+1, n+1)):
            y[n] -= a[k]*y[n-k]
            
        y[n] /= a[0]
        
    return y
```

# Oppsummering Frekvensrespons Digitale Filtre

* ***Frekvensresponsen*** til et filter med impulsrespons $h[n]$ er impulsresponsens fourietransformerte:
$$H\left(\hat{\omega} \right) = \text{DTFT}(h[n]) = \sum_{n=0}^{\infty} h[n]\cdot e^{-j\hat{\omega}\cdot n}$$
* Gir forholdet mellom inngangssignal og utgangssignal til et filter i frekvensplanet.
$$Y\left(\hat{\omega} \right) = X\left(\hat{\omega} \right)\cdot H\left(\hat{\omega} \right) \leftrightarrow H\left(\hat{\omega} \right) = \frac{Y\left(\hat{\omega} \right)}{X\left(\hat{\omega} \right)}$$
* Forteller hvordan filteret vil påvirke en hvilken som helst sinussekvens:
\begin{align}
x[n] &=  A\cdot \cos(\hat{\omega}_0 \cdot n + \phi)\\
& \downarrow \\
y[n] &= \left| H\left(\hat{\omega} \right)\right| \cdot A\cdot \cos\left(\hat{\omega}_0 \cdot n + \phi + \angle H\left(\hat{\omega} \right)\right)
\end{align}

# Frekvensrespons førsteordens IIR-filter

* Vi kan fortsatt finne frekvensresponsen til ved å fouriertransformere impulsresponsen $h[n]$.
\begin{align}
y[n] &= b_0 \cdot x[n] - a_1 \cdot y[n-1]\\
&\downarrow \\
h[n] &= b_0\cdot u[n] \cdot (-a_1)^n \\
&\downarrow \\
H\left(\hat{\omega}\right) &= \frac{b_0}{1+a_1\cdot e^{-j\hat{\omega}}}, \ \ \ |a_1| < 1
\end{align}
* DTFT av geometrisk rekke gitt i tabell over vanlige fouriertransformasjonspar.

# Frekvensrespons via differanseligning



\begin{align}
DTFT\left(\sum_{k=0}^{N} a_k \cdot y[n-k]\right) &= DTFT\left(\sum_{k=0}^{M} b_k \cdot x[n-k] \right)\\
&\downarrow \\
& \text{DTFT}(x[n-k]) = e^{-j \hat{\omega} \cdot k} \cdot X(\hat{\omega})\\
& \downarrow \\
\sum_{k=0}^{N} a_k \cdot Y\left(\hat{\omega}\right)\cdot e^{-j\hat{\omega}\cdot k} &=\sum_{k=0}^{M} b_k \cdot X\left(\hat{\omega}\right)\cdot e^{-j\hat{\omega}\cdot k}\\
Y\left(\hat{\omega}\right)\cdot \left(\sum_{k=0}^{N} a_k \cdot e^{-j\hat{\omega}\cdot k}\right) &= X\left(\hat{\omega}\right)\cdot \left(\sum_{k=0}^{M} b_k \cdot e^{-j\hat{\omega}\cdot k}\right)\\
H\left(\hat{\omega}\right) &= \frac{Y\left(\hat{\omega}\right)}{X\left(\hat{\omega}\right)} \\ \\
&= \frac{\sum_{k=0}^{M} b_k \cdot e^{-j\hat{\omega}\cdot k}}{\sum_{k=0}^{N} a_k \cdot e^{-j\hat{\omega}\cdot k}}
\end{align}

# Stabilitet

* Med rekursive filter åpnes døren for å lage *ustabile* system. 
* I et ustabilt system vil *ikke* impulsresponsen $h[n]$ / stegresponsen $s[n]$ konvergere mot en bestemt verdi.
* Å sikre at det digitale filteret er stabilt er et sentralt aspekt i design av IIR filtre.

# Refleksjoner

* Ved å lage et *rekursivt* filter, kan selv enkle førsteordens filtre produsere en langvarig impulsrespons.
* Et IIR-filter med relativt lav filterorden kan, dersom vi velger filterkoeffisienter med omhu, oppfylle de samme kravene for amplitduerespons som et FIR filter med ***mye*** høyere filterorden.
* Ulempen er at IIR filtre ***ikke*** har lineær faserespons i passbåndet. 
    * Konsekvensen av dette er *tidsforskyvningen* for hver av frekvenskomponentene til det filtrerte signalet gitt i antall sampler ikke vil være konstant.
    * Dette har potensiale til å forårsake uønsket distorsjon av f.eks. et lydsignal.
* IIR-filtre har potensialet til å være ustabile.
* Det vil være *svært* matematisk tungvindt å designe et IIR-filter ved å først identifisere ønsket impulsrespons (slik som med FIR filtre)
