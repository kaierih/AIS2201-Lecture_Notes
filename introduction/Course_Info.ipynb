# Why learn digital signal processing?

### *Any suggestions?*

##### Notes:
Digital signal processing is a broad field of study with a surprisingly wide range of applications, ranging from wireless communication systems to mapping of structural dynamics and even analysis of stock exchange data. As aspiring automation engineers, there are two main areas in which digital signal processing are likely to be of use:

1. **Signal analysis in troubleshooting.**<br>
Making use of tools such as spectral representation of a signal can help identify sources of disturbance and/or failure in an automated system relying on sensor measurements.
2. **Extracting desired information from imprecise/indirect measurements.**<br>
Digital filters and fourier transforms are commonly used to aid in estimating properties which are present in have to be calculated by analyzing a measured signal. 

In fact, it is highly likely that you have already made use of some basic signal processing tools earlier in your studies. A common first encounter with digital signal processing for instance is writing a program to calculate a moving average of sensor values to reduce measurement noise. Furthermore, if you have taken the course AIS2102 - Dynamical Systems, you would have spent three weeks learing about [discretization](https://en.wikipedia.org/wiki/Discretization), which covers the use of digital filters to emulate a dynamical system in discrete time. 

This course aims to provide the necessary understanding to make use of the fundamental signal processing tools which are among the most universally used. The teaching activities take a problem-oriented approach to introducing central concepts, and covers the systems theory necessary to design and analyze a digital signal processing system, as well as implementation of digital signal processing systems on a microprocessor.

# What *is* signal processing?

1. ***Signal processing*** focuses on analyzing or manipulating a *signal*, often with the purpose of extracting specific information.

2. A ***Signal*** may be any measurable function containing information, which varies with time/location etc...

3. Signal processing conducted on a computer or other digital processing unit is known as ***Digital Signal Processing***.

## Python script example: *Sinusoid*

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl

t = np.linspace(0, 1, 501)
x_t = np.cos(6*np.pi*t)

plt.close(1); plt.figure(1, figsize=(12, 5))
plt.plot(t, x_t, color='tab:orange')
```
